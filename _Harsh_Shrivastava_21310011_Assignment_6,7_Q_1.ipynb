Given Initial Point A (0.4,0.06,0.1);
Final Point, B (0.4,0.01,0.1)

Using Polymnomial Trajectory method to find the trajectory of end effector

```python
# Let's find the x,y,z position, velocity, acceleration
x = np.array([])
vx = np.array([])
ax = np.array([])

y = np.array([])
vy = np.array([])
ay = np.array([])

z = np.array([])
vz = np.array([])
az = np.array([])

A = np.array([])
for t in range(0,101,1):
    x = np.append(x,0.4)
    vx = np.append(vx,0)
    ax = np.append(ax,0)
    
    y = np.append(y,0.06 -0.5*((t/100)**3)+0.75*((t/100)**4)-0.3*((t/100)**5))
    vy = np.append(vy,-3*0.5*((t/100)**2)+4*0.75*((t/100)**3)-5*0.3*((t/100)**4))
    ay =np.append(ay,-6*0.5*((t/100))+12*0.75*((t/100)**2)-20*0.3*((t/100)**3))
    
    z = np.append(z,0.1)
    vz = np.append(vz,0)
    az = np.append(az,0)

    A = np.append(A,t/100)

plt.figure(1)
#plt.plot(A,x,'b',A,y,'g',A,z,'r')
plt.plot(A,y,'g')
plt.xlabel('Time')
plt.ylabel('Position')
plt.legend(['X','Y','Z'])

plt.figure(2)
plt.plot(A,vx,'b',A,vy,'g',A,vz,'r')
plt.xlabel('Time')
plt.ylabel('Velocity')
plt.legend(['Vx','Vy','Vz'])

plt.figure(3)
plt.plot(A,ax,'b',A,ay,'g',A,az,'r')
plt.xlabel('Time')
plt.ylabel('Acceleration')
plt.legend(['Ax','Ay','Az'])
```
